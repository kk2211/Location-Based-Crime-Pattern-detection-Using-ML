```python
import pandas as pd
```

```python
df = pd.read_csv('alyien_news_data.csv')
df=df.drop(['time', 'description', 'url', 'summary'], axis=1)
df.head()
```

```python
import nltk
from nltk.tag.stanford import StanfordNERTagger

jar = './stanford-ner.jar'
model = './english.all.3class.distsim.crf.ser.gz'

def findLocations(sentence):
    locations = []
    ner_tagger = StanfordNERTagger(model, jar, encoding='utf8')
    words = nltk.word_tokenize(sentence)
    lst = ner_tagger.tag(words)
    
    i=0
    while i<len(lst):
        if lst[i][1]=='LOCATION':
            s = lst[i][0]
            i+=1
            if i==len(lst):
                break
            while lst[i][1]=='LOCATION':
                s = s+' '+lst[i][0]
                i+=1
                if i==len(lst):
                    break
            if s.lower() not in locations:
                locations.append(s.lower())
        else:
            i+=1
        
    return locations    
```

```python
from word2number import w2n
def findAge(content):
    if 'year-old' in content:
        index = content.index('year-old')
        if index-7<0:
            age = content[:index+8].split()
        else:
            age = content[index-7:index+8].split()
        if len(age)!=0:
            if age[-1]=='year-old':
                if '-' in age[-2]:
                    s = str(age[-2][:-1])
                else:
                    s = str(age[-2])
            else:
                s = age[-1].split('-')[0]
            if s=='a':
                s = 1
            s = str(s).split('.')[-1]
            return w2n.word_to_num(str(s))
    else:
        return -1
```

```python
contents = ' '.join(list(df['content']))
locations = findLocations(contents)
```

```python
f=open('loc.txt', 'r')
s=f.read()
```

```python
loc = s.split('\n')[3:]
delhi_locs = []
for i in loc:
    if i!='LOCALITY':
        delhi_locs.append(i.lower())
```

```python
all_locations = list(set(delhi_locs).intersection(locations))
len(all_locations)
```

```python
business = ['businessman', 'jeweller', 'jeweler', 'shop owner', 'property dealer']
```

```python
d = {}
```

```python
c=0
for i in range(len(df)):
    print(i, c)
    article = df['content'].iloc[i]
    locs = findLocations(article)
    date = df['date'].iloc[i]
    headline = df['title'].iloc[i]
    crime_type = df['type'].iloc[i]
    businessman = 0
    for j in business:
        if j in article:
            businessman = 1
            break
    age = findAge(article)
    
    for i in locs:
        if i in all_locations:
            c+=1
            flag = 1
            lst = d.get(i, [[],[],[],[],[],[]]) # date, headline, crime_type, article, age, businessman
            lst[0].append(date)
            lst[1].append(headline)
            lst[2].append(crime_type)
            lst[3].append(article)
            lst[4].append(age)
            lst[5].append(businessman)
            d[i] = lst
            
```

```python
rows = []
header = ['location', 'date', 'headline', 'crime_type', 'article', 'age', 'businessman']
rows.append(header)
for i in d.keys():
    row = [i]
    lst = d[i]
    row.append('\t'.join(lst[0]))
    row.append('\t'.join(lst[1]))
    row.append('\t'.join(lst[2]))
    row.append('\t'.join(lst[3]))
    row.append('\t'.join(repr(n) for n in lst[4]))
    row.append('\t'.join(repr(n) for n in lst[5]))
    rows.append(row)
    
```

```python
import csv
data = open('news_data.csv', 'w')
writer = csv.writer(data)
writer.writerows(rows)
data.close()
```
